## Let's try to scrap some earthquake infomation from the CESMD Strong-Motion Data Set

### Here we will scrape the following website:
<html>https://www.strongmotioncenter.org/cgi-bin/CESMD/archive.pl?Archives="Year"</html> to earthquake record since 2023

```python
HTML('https://www.strongmotioncenter.org/cgi-bin/CESMD/archive.pl?Archives=2023') #opens the webpage
```

```python
url = urllib.request.urlopen('https://www.strongmotioncenter.org/cgi-bin/CESMD/archive.pl?Archives=2023')
content = url.read()
print (content[0:1000])
```

```python
soup = BeautifulSoup(content, 'lxml')
print(soup)
```

```python
for year in range(2023,2026):
    url = urllib.request.urlopen('https://www.strongmotioncenter.org/cgi-bin/CESMD/archive.pl?Archives=' + str(year))
    html_content = url.read()

    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the table
    table = soup.find('table', {'id': 'eqdata'})

    # Extract the table headers
    headers = [header.get_text(strip=True) for header in table.find_all('th')]

    # Extract the table rows
    records = []
    for record in table.find_all('tr')[1:]:  # Skip the header row
        dataEntry = record.find_all('td')
        record_data = [col.get_text(strip=True) for col in dataEntry]
        records.append(record_data)

    # Print the extracted data
    #print(headers)
    #for record in records:
    #    print(record)
```

```python
eqRecords = pd.DataFrame(records, columns=headers)
```
