### Convolution layers
* Consider a 1-D case: x (x0-x4) and w (w1-w3) 

* Sparse connectivity: not every input unit is connected to every output unit.

* Weight sharing: the network’s weights are each shared between multiple connections.

![alt text](./img/con06.png "convLayer")




* Convolution layers normally compute multiple feature maps $z_1$, . . . , $z_M$. The input layers also consist of multiple feature maps $x_1$, . . . , $x_D$
* These could be different color channels of an RGB image, or feature maps computed by another convolution layer. There is a separate filter $W_{ij}$ associated with each pair of an input and output feature map. The activations are computed as follows:

$$z_i = \sum_j{x_j * w_{ij}} $$
$$h_i = \phi(z_i)$$

## Considere the case:

* 32 input feature maps and 16 output feature maps all of size 50 × 50, and the filters are of size 5 × 5. 
* The number of weights for the convolution layer is 5 × 5 × 16 × 32 = 12,800.
* The number of connections is approximately 50 × 50 × 5 × 5 × 16 × 32 = 32 million.

* What is the number of connections and the number of weights required for a fully connected layer with the same set of units? 


(32 × 50 × 50) × (16 × 50 × 50) = 3.2 billion.

* The convolutional structure reduces the number of connections by a factor of 100 and the number of weights by almost a factor of a million!

* This means convolution layers don’t increase the representational capacity, relative to a fully connected layer with the same number of input and output units. But they can reduce the numbers of weights and connections.
